```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split

import os
os.chdir(r"C:\Users\Excalibur\Desktop\VERİ BİLİMİ\Makine Öğrenmesi\verisetleri")
```

### ONE HOT ENCODING

```python
one_hot = OneHotEncoder(handle_unknown="ignore")
# handle_unknown="ignore" : bilinmeyen bir değer geldiğinde umursama diyoruz.

one_hot_cinsiyet = one_hot.fit_transform(df.cinsiyet.values.reshape(-1,1)).toarray()
# toarray() : numpy array dizisine çeviriyoruz.
```

```python
one_hot.categories_
```

```python
one_hot_cinsiyet  # 1 sütun erkek, 2.sütun kadın
```

```python
one_hot_df = pd.DataFrame(one_hot_cinsiyet, columns= one_hot.categories_)
```

```python
one_hot_df
```

```python
df2 = df.join(one_hot_df)
df2.head()
```

```python
df2.drop("cinsiyet", axis=1, inplace=True)
df2.head()
```

```python
X = df2.drop("kredi",axis=1)
y = df2.kredi
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size = 0.33,
                                                    random_state = 42)
```

```python
X_train # verisetinden rastgele %33 lük bir alanı aldı
```

**Ölçeklendirme**

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

**Modelleme Aşaması**

```python
model = LogisticRegression(random_state=0)
model.fit(X_train, y_train) # eğitimi gerçekleştirdik
```

**Tahmin Sonuçları**

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test.values
```

**Performans Değerlendirmesi**

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
# sınıflandırmada performans değerlendirmelerinde kullanılan metriklerdir.
# accuracy_score : doğruluk skoru yüzde verir. ne kadarını doğru bildik.
```

```python
confusion_matrix(y_true = y_test, y_pred = y_pred)
```

* 78 ve 33 doğru tahminler
* 19 ve 2 yanlış tahminlerdir.

```python
classification_report(y_true = y_test, y_pred = y_pred)
```

```python
accuracy_score(y_true = y_test, y_pred = y_pred) # %84 oranında kredi verilip verilmediğini tahmin etmiş olduk
```
