# ÇİZGİ(Line) GRAFİĞİ

```python
fig = plt.figure() # grafik nesnesi oluşturuyoruz.
ax = plt.axes() # grafik alanı oluşturuyoruz.
# x ekseni için 0 ile 10 arasında 100 değer alalım.
x = np.linspace(0,10,100)
y = np.sin(x)
# Bunların grafiğini çizelim.
ax.plot(x,y) 
```

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

```python
plt.plot(x,np.sin(x-1), color="blue")
plt.plot(x,np.sin(x-2), color="r")
plt.plot(x,np.sin(x-3), color="0.45") # "0.75" grafiğin grilik değerini girebiliriz.
plt.plot(x,np.sin(x-4), color="#854596")
```

```python
plt.plot(x,x+1, "r-")
plt.plot(x,x+2, "b--")
plt.plot(x,x+3, "g-.")
plt.plot(x,x+4, "k:") # k: siyah
```

**linestyle :** Çizgi sitili için kullanılır.

```python
plt.plot(x,x+1, linestyle="solid")  # "-" düz çizgi için
plt.plot(x,x+2, linestyle='dashed')  # "--" kesikli çizgi
plt.plot(x,x+3, linestyle='dashdot') # "-." şeklinde
plt.plot(x,x+4, linestyle='dotted') # ":" bu şekilde oluşturulur nokta şeklinde 
```

```python
plt.plot(x,x+2, 'go--', linewidth=2, markersize=12)
plt.plot(x,x+3, color='green', marker='o', linestyle='dashed',
         linewidth=2, markersize=12)
```

Eksenlerin aralıklarını belirleyebilmek için **xlim** ve **ylim** metodları kullanılır.

```python
plt.plot(x,np.sin(x))
plt.xlim(-5,20)
plt.ylim(-3,3)
```

**axis()** metodu ile de x ve eksenleri için değer girilebilir.

```python
plt.plot(x,np.sin(x))
plt.axis([-5,10,-3,3]) # önce x eksen aralığı sonra da y eksen arallığı girilir.
```

axis metodunun **tight** gibi otomatik ayarlamalara izin veren fonksiyonları vardır.

```python
plt.plot(x,np.sin(x))
plt.axis("tight") # grafik değerlere göre otomatik olarak ayarlanmış oldu
plt.title("Bir sinüs eğrisi")
plt.xlabel("x değerleri")
plt.ylabel("y değerleri")
```

* Çizgileri kendimiz isimlendirmek istersek plot metodu içinde label argümanı kullanılır.

```python
plt.plot(x,x+2, "-r", label="falan")
plt.plot(x,x+4, ":b", label="filan")
plt.legend()
```

# SAÇILIM(Scatter) GRAFİĞİ

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
plt.style.use("seaborn-whitegrid")
fig = plt.figure()
ax=plt.axes()
plt.plot(iris.sepal_length,iris.sepal_width,"ro", markersize=10) # saçılım grafiğimi plot ile çizdirebiliriz.
# markersize noktaların boyutunu değiştirir.
```

```python
plt.scatter(iris.sepal_width,iris.petal_width, s=75, c="m", 
            marker="v",edgecolors="black", linewidths=1, alpha=1) # saçılım grafiğini çizer
# s  argümanı noktaların boyutunu ayarlar.
# c  argümanı noktaların rengi için kullanılır.
# marker  noktaların şekli için kullanılır.
# linewidths  noktaların çerçeve kalınlığı için kullanılır.
# edgecolors  noktaların çerçeve rengi için kullanılır.
# alpha  noktaların görünürlüğünü ayarlar.
```

scatter metodunun plot metodundan farkı noktaların özelliklerini ayrı ayrı almasıdır.

* Her bir noktanın griliğini ayarlamak isteyebiliriz. Bunun için bir renk değişkeni oluşturalım. Renk değişkenini numpy random metodundaki randint metodundan sıfır ile 10 arasında 150 tane sayı üretelim.

```python
renk = np.random.randint(0,10,150)
plt.scatter(iris.sepal_width,iris.petal_width,s=100, c=renk, alpha=1)
```

Noktaların rengini ayarlayabiliriz bunun için **"cmap"** kullanabiliriz. Girdiğimiz değerin ilk harfi büyük sonuna da "s" harfi konulur. (Reds,Blues gibi)

```python
renk = np.random.randint(0,10,150) # kırmızının farklı tonları
plt.scatter(iris.sepal_width,iris.petal_width,s=100, c=renk, cmap="Reds", alpha=1)
```

```python
# Daha canlı renkler kullanmak istersek cmap="viridis" yazabiliriz.
renk = np.random.randint(0,10,150) 
plt.scatter(iris.sepal_width,iris.petal_width,s=100, c=renk, cmap="viridis", alpha=1)
```

Sayı değerlerini renkler ile gösteren çubuk ekleyebiliriz. Bunun için **colorbar** metodu kullanabiliriz. **colorbar**'a **set_label()** ile isim verebiliriz.

```python
renk = np.random.randint(0,10,150) 
plt.scatter(iris.sepal_width,iris.petal_width,s=100, c=renk, cmap="viridis", alpha=1)
cbar = plt.colorbar()
cbar.set_label("Seviyeler")
```

Noktaların  boyutlarını da farklı yapabiliriz."boyut" isminde bir değişken ıluşturup numpy.random.randit fonksiyonunu kullanarak 50 ile 200 arasında 150 tane sayı üretip s argümanına atayalım.

```python
plt.style.use("fivethirtyeight")
boyut = np.random.randint(50,200,150)
renk = np.random.randint(0,10,150) 
plt.scatter(iris.sepal_width,iris.petal_width,s=boyut, c=renk, cmap="viridis", alpha=1)
cbar = plt.colorbar()
cbar.set_label("Seviyeler")
# eksenleri isimlendirebiliriz.
plt.xlabel("sepal genişliği")
plt.ylabel("petal genişliği")
# Grafiğe başlık verebiliriz.
plt.title("İris çiçeği sepal en & petal en")
```

**?** Peki hangisini seçeceğiz? Scatter mi yoksa Plot metodunu mu?
* Küçük örneklemlerde ikis arasında fark yok ama büyük örneklemlerde **Plot** metodu daha iyidir.Çünkü daha hızlıdır ve her bir gözlem değerini daha net gösterir.

# BAR GRAFİĞİ

* x ekseninde kategorik değişken var ise bar grafiği çizmek daha mantıklıdır.

```python
plt.style.use("fivethirtyeight")
```

```python
mat = np.random.randint(50,100,7)
fen = np.random.randint(50,100,7)
ing = np.random.randint(50,100,7)
sinif = list("ABCDEFG")
```

```python
plt.bar(sinif, mat , label="mat ortalama")
plt.bar(sinif, fen , label="fen ortalama")
plt.bar(sinif, ing , label="ing ortalama")
plt.legend()
plt.title("Sınıfların Ders Ortalamaları")
plt.xlabel("Sınıflar")
plt.ylabel("Ortalamalar")
```

* Karşımıza üst üste çizilmiş barlar geldi. Verilerin ayrı ayrı barlarını görmek istersek;

* Barların genişliği varsayılan olarak 0.8 gelir. Bunu **"width"** ile ayarlayabiliriz.

```python
x_index = np.arange(len(sinif)) # Öncelikle verileri indexliyoruz.
en = 0.25 # barların üst üste gelmesini önlemek için bir değişken tanımlıyoruz.
plt.bar(x_index-en, mat, width=en , label="mat ortalama") # her bir grafiğin indexinin farklı olması için bu değişkeni metodlarda kullanıyoruz.
plt.bar(x_index, fen, width=en , label="fen ortalama")
plt.bar(x_index+en, ing, width=en , label="ing ortalama")
plt.legend()
plt.title("Sınıfların Ders Ortalamaları")
plt.xlabel("Sınıflar")
plt.ylabel("Ortalamalar")
```

 * x ekseninde sayılar görünüyor bunu düzeltelim.

```python
x_index = np.arange(len(sinif)) 
en = 0.25

plt.bar(x_index-en, mat, width=en , label="mat ortalama")
plt.bar(x_index, fen, width=en , label="fen ortalama")
plt.bar(x_index+en, ing, width=en , label="ing ortalama")
# bbox_to_anchor legendin konumunu belirler
plt.legend(bbox_to_anchor=(1, 0., 0.4, 0.3)) # aşağı,sol,sağ,yukarı
plt.xticks(ticks=x_index, labels=sinif) # x ekseninde sayılar yerine sinif etiketleri olacaktır.

plt.title("Sınıfların Ders Ortalamaları")
plt.xlabel("Sınıflar")
plt.ylabel("Ortalamalar")
```

* Bilimsel çalışmalarda hatalarda düşünülerek ölçüm yapılır. Hata Çubukları Grafiğini inceleyelim.

```python
x = np.linspace(0,10,30)
y = np.random.randint(0,5,30)
# x ve ye için saçılım grafiği
plt.plot(x,y,"bo") # bu noktalara hata barları ekleyelim.
plt.errorbar(x,y,yerr=1,fmt="b.", ecolor="m", elinewidth=2) 
# yerr : y ekseni boyunca yani dikey hata çubukları için y error un kısaltması yerr kullanılır. Uzunluğu 1 olsun.
# fmt : renk ve görünümü ayarlar. mor ve nokta şeklinde ayarladık
# ecolor : hata çubukları rengi için kullanılır.
# elinewidth : hata çubuklarının genişliğini belirler.
```

> Bar ile hata çubuklarını beraber gösteren grrafiği çizdirelim.

```python
from sklearn.datasets import load_iris # iris veri setini sklearn ile import edelim.
iris = load_iris()
```

```python
ort =np.mean(iris.data, axis=0) # sütun boyunca hesaplama
sd = np.std(iris.data, axis=0)
```

```python
ort
```

```python
sd
```

```python
aralik = range(4) # iris veri setinde 4 sayısal değişken olsuğu için 4 yazdık.
aralik
```

```python
# Hata çubukları ile 2 tane bar grafiği çizdirelim. Biri yatay(h) diğeri dikey(v)
plt.subplot(1,2,1) # 2 grafik kullanacağımız için 1 süzlemde 2 alan ve 1. grafik
plt.barh(aralik, ort, color="b", xerr=sd, alpha=0.7, align="center",height=0.5) # barh yatay çizim
plt.title("Yatay Bar Grafiği")
plt.yticks(aralik, iris.feature_names) # etiketler iris veri setindeki değişkenleri isim olarak alsın


plt.subplot(1,2,2)
plt.bar(aralik,ort,color="m",yerr=sd,alpha=0.7, align="center",width=0.5)
plt.title("Dikey Bar Grafiği")
plt.xticks(aralik, aralik)
```

> barların genişliklerini ayarlayabiliriz. yataş graafik için **height** dikey grafik için **width** kullanılır. varsayılan 0.8 dir

# HİSTOGRAM GRAFİĞİ

* Bir değişkenin dağılımını görmek için histogramlar kullanılır. Histogramlar verinin değeri yerine frekansları gösterir.

```python
# normal dağılımdan bir veri seti oluşturalım
veri = np.random.randn(1000)
plt.hist(veri, edgecolor="black")
plt.title("Histogramlar")
plt.xlabel("Değerler")
plt.ylabel("Frekankslar")
```

```python
veri = np.random.randn(1000)
plt.hist(veri, edgecolor="black", bins=10) # bins=10 x ekseni 10 parçaya bölünür. 
plt.title("Histogramlar")
plt.xlabel("Değerler")
plt.ylabel("Frekankslar")
```

* Dikdörtgenler kendi alanlarına kaç tane değer düştüğünü yani frekansları gösteriyor.
* histogranların kenar çizgilerini görmek istersek **edgecolor** argümanı kullanılır.

```python
# normal dağılımdan bir veri seti oluşturalım
veri = np.random.randn(1000)
bins =[-3,-2,-1,0,1,2,3] # bins argümanına dikdörtgenin sınırlarını belirleyen değerlerde girebiliriz.
plt.hist(veri, edgecolor="black", bins=bins) 
plt.title("Histogramlar")
plt.xlabel("Değerler")
plt.ylabel("Frekankslar")
```

> Verinin ortanca değerlerini gösteren bir çizgi ekleyebiliriz.

```python
veri = np.random.randn(1000)

ortanca = np.median(veri)
plt.axvline(ortanca, color="yellow")

plt.hist(veri, edgecolor="black", bins=10) 
plt.title("Histogramlar")
plt.xlabel("Değerler")
plt.ylabel("Frekankslar")
```

```python
# Bir kaç değişkenin histogramlarını karşılaştırabiliriz.
x1 = np.random.normal(0,0.8,1000) # ort, std, gözlem sayısı 
x2 = np.random.normal(-3,2,1000)
x3 = np.random.normal(2,1,1000)
```

```python
# Şimdibu değişkenlerin histogram grafiklerini çizdirelim.
plt.hist(x1, histtype="stepfilled", alpha=0.5)
plt.hist(x2, histtype="stepfilled", alpha=0.5)
plt.hist(x3, histtype="stepfilled", alpha=0.5)
#  histtype : {'bar', 'barstacked', 'step', 'stepfilled'}, default: 'bar'
```
