# Matplotlib
Matplotlib, Python'da statik, hareketli ve etkileşimli görselleştirmeler oluşturmak için kullanılan kapsamlı bir kütüphanedir.
 
2 ve 3 boyutlu çizimler yapmamızı sağlar. Matplotlib genelde 2 boyutlu çizimlerde kullanılırken, 3 boyutlu çizimlerde başka kütüphanelerden yararlanılır.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
veri = np.arange(10)
```

* Grafikler için interaktif pencere kullanalım. **interaktif**, grafik açıkken üzerinde değişiklik yapılabilir anlamına gelmektedir.

```python
%matplotlib notebook 
# Çizeceğimiz grafikler farklı bir pencere de karşımıza gelecektir.
```

**plt.plot():** Grafiğin ana komutunu oluşturur. Bu komutun içine ilk önce x ekseni, sonrasında y ekseni yazılır. Grafikteki temel bazı değişkenleri değiştirmek istersek ilgili kodu buraya yazmamız gerekir. Örneğin grafiğin rengi, çizgi kalınlığı, aralıklar vs.

   * **figsize:** Grafiğin boyutunu belirler. Yaygın olarak **"plt.figure(figsize=(8,6))"** şeklinde kullanılır. Farklı kullanımları da mevcuttur. Parantez içindeki ilk yazılan değer grafiğin eni, ikincisi boyudur.
   * **plt.xlabel():** x ekseninin başlığını oluşturur.
   * **plt.ylabel():** y ekseninin başlığını oluşturur.
   * **plt.title():** Grafiğin başlığını oluşturur.
   * **plt.show():** Grafiği çizdirmemizi sağlar. Bu kodu yazmazsak grafik çizilir fakat bazen ekranda gözükmeyebilir.

```python
plt.figure(figsize=(12,6))
plt.plot(veri)
plt.title("Grafik Başlığı")
plt.xlabel("x ekseni")
plt.show()
```

Matplotlib kütüphanesinin listeler ile çalışması sınırlıdır. Genellikle sayısal hesaplamalar için numpy kullanılır.

```python
d = np.arange(0.,5.,0.2)
d
```

* Bu diziyi farklı formatlarda çizdirelim.

    * **r-- :** kırmızı ve kesikli çizgi 
    * **bs  :** mavi ve kare
    * **g^  :** yeşil ve üçgen
    * **k-- :** siyah ve kesikli çizgi

```python
plt.plot(d,d,"r--",d,d**2,"bs",d,d**3,"g^")
```

> Grafiği **plt.savefig()** ile bilgisayara kaydedebiliriz.

**plt.savefig()** yapısında 2 önemli özellik vardır.
1. **dpi=** grafiğin görüntü kalitesini ayarlamak için kullanılır.
2. **bbox_inches="tight"** grafiğin çevresindeki beyaz alanın kesilmesini sağlar.

Örneğin " png " formatında kaydedelim.

```python
plt.plot(d,d,"r--",d,d**2,"bs",d,d**3,"g^")
plt.savefig("grafiğim1.png", dpi=500, bbox_inches="tight")
```

* Grafikler **figure** nesnesi ile dizayn edilir. 

* Birden fazla alt grafik oluşturmak için **add_subplot()** metodu kullanılır.

```python
fig = plt.figure()
# boş bir grafik oluşturduk.
x1 = fig.add_subplot(2,2,1) 
# "2 satır ve 2 sütun kaplayan 4 grafik alanından 1. si x1 dir" diyoruz.
data = np.random.randn(50).cumsum()
# normal dağılımdan rasgele 50 veri üretip bunun kümülatif toplamını elde ediyoruz.
# Kümülatif toplam; birikmeli, gittikçe artan, birbirine eklenen toplam
plt.plot(data,"k--") # 1. grafik alanına grafiği çizdiriyoruz.
```

```python
x2 = fig.add_subplot(2,2,2) # 2. grafik alanı oluşturuldu
x3 = fig.add_subplot(2,2,3) # 3. grafik alanı oluşturuldu
```

```python
# Bu alanları dolduralım.
x2.plot(data,"r-")
x3.plot(data,"bo")
```

```python
x = [1,2,3,4,5]
y = [10,11,12,13,14]
fig = plt.figure() # grafik nesnesi oluşturduk.
ax = fig.add_subplot(1,1,1) # eksenleri oluşturduk
ax.plot(x,y,"r--")
plt.plot(data,"ko--") 
# Ana veri noktalarını gösteren işaretler ekleyebiliriz.
# k : kırmızı , o : ana noktaları bekirden içi dolu daire, -- : ana noktalardan geçen çizgi
```

* **legent** grafik elemanlarını tanımlayan bileşendir. 
legent kullanmanın en kolay yolu plot ın her bir parçasına label özelliğini eklemektir.

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
veri1 = np.random.randn(1000).cumsum()
veri2 = np.random.randn(1000).cumsum()
veri3 = np.random.randn(1000).cumsum()
ax.plot(veri1,"k",label="bir")
ax.plot(veri2,"b--", label="iki")
ax.plot(veri3,"r.",label="uc")
ax.legend(loc="best")
```

* **parse_dates=True** iletirseniz, dizinin **datetime** türünde olduğunu varsayar.
**parse_dates :** boolean veya ints listesi veya adlar veya listeler veya dict listesi, varsayılan False boolean. True ise -> dizini ayrıştırmayı deneyin. 
* ints veya adların listesi. örneğin [1, 2, 3] -> ise 1, 2, 3 sütunlarının her birini ayrı bir tarih sütunu olarak ayrıştırmayı deneyin. 
* listeler listesi. örneğin [[1, 3]] -> ise 1 ve 3 numaralı sütunları birleştirin ve tek bir tarih sütunu olarak ayrıştırın. 
* dict, örneğin {'foo' : [1, 3]} -> sütun 1, 3'ü tarih olarak ayrıştır ve sonucu 'foo' olarak çağır Not: iso8601 formatlı tarihler için bir hızlı yol mevcuttur.

* Facebook 'un kapanış fiyatlarını gösteren grafiği çizdirelim.

```python
k_fiyat = veri["Close"]
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
k_fiyat.plot(ax=ax,style="k-")
ax.set_title("Facebook borsa kapanış fiyatları")
```

* Grafikteki en küçük değeri grafik üzerinde OK işareti ile gösterelim.

```python
k_fiyat["2018-12-24"] # görüldüğü üzere min değer nu tarihtedir.
```

```python
tarih = "2018-12-24"
yazi = "Minimum Değer"
```

> **annotate** metodu ile grafiğe ok ve yazı eklenir.
Bu metotda ki **xy** argümanı lokasyonu, **xytext** argümanı ise yazının lokasyonu için kullanılır. **arrowprops** ok için gerekli opsiyondur.

```python
# xytext=(tarih,150) -> x ekseni için tarih, y ekseni için 150 değerini alıyoruz.
ax.annotate(yazi, xy=(tarih,125), xytext=(tarih,150), arrowprops=dict(facecolor="red"))
# grafiğe baktığımızda grafikte min değeri gösteren kırmızı bir ok göreceğiz.
```

```python
plt.savefig("min_deger_grafiği.pdf", dpi=500,bbox_inches="tight")
# pdf türünde grafiği veri setinin olduğu dosyaya kaydettik.
```

* Çözünürlük varsayılan olarak 100 gelir. **dpi** özelliği ile çözünürlüğü ayarlayabiliriz.
* **bbox_inches="tight** argümanı grafiğin etrafındaki boş argümanları keser.
